````python
#importing necessary libraries
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
````

````python
#loading dataset of Canada's per capita income
df = pd.read_csv('/content/canada_per_capita_income.csv')
df
````

````python
#ploting graph of per capita of Canada with year to visualize
%matplotlib inline
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='o')
````

````python
#Spliting dataset to train model
new_df = df.drop('per capita income (US$)',axis='columns')
per_capita_income = df.drop("year", axis=1)
````

````python
#training Linear Regression model
reg = linear_model.LinearRegression()
reg.fit(new_df,per_capita_income)
````

````python
#finding coefficient to calculate 2020's per capita for Canada
reg.coef_
````

````python
#finding intercept to calculate 2020's per capita for Canada
reg.intercept_
````

## **Equation for Simple Linear Regression**
**Y = m * X + b (m is coefficient and b is intercept)**
